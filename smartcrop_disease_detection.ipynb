```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from google.colab import files
```

```python
print("Upload your trained model file (model.h5)")


uploaded = files.upload()


model_path = list(uploaded.keys())[0]


model = load_model(model_path)
print("Model loaded successfully!")
```

```python
 print("Upload all test images")

 uploaded_imgs = files.upload()

 labels_dict = {0: "Healthy", 1: "Diseased"}
 y_true = []   # true labels
 y_pred = []   # predicted labels

 for fn in uploaded_imgs.keys():
     img = load_img(fn, target_size=(224, 224))
     plt.imshow(img)
     plt.axis("off")
     plt.title(fn)
     plt.show()
     x = img_to_array(img)
     x = np.expand_dims(x, axis=0)
     x = preprocess_input(x)
     prediction = model.predict(x)[0][0]
     if prediction > 0.5:
      pred_class = 1
      confidence = prediction
     else:
        pred_class = 0
        confidence = 1 - prediction
        y_pred.append(pred_class)
        true_label = int(input(f"Enter TRUE label for {fn} (0=Healthy, 1=Diseased): "))
        y_true.append(true_label)
        print(f"{fn} → Predicted: {labels_dict[pred_class]} (Confidence: {confidence:.2f}) | True: {labels_dict[true_label]}")
```
